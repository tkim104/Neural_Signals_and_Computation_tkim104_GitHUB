Load up the calcium image video in the file: TEST MOVIE 00001-small.tif. This file represents the raw fluorescence video in TIF format, but without the motion. One of the most common ways that people analyze functional data is to try to identify individual cells in the data in a “summary image”. A summary image condenses an entire video sequence into a single image, wherein each pixel represents a summary of the entire time-trace at the location of that pixel in the video.

## Part A
Plot the “mean”, “median”, and “variance” images for the data provided. What do you notice about the visible cells in each of these?

```python
import plotly
import plotly.graph_objects as go
import tifffile
from google.colab import drive
from PIL import Image
import numpy as np

# Find the tif file in google drive
drive.mount('/content/drive')
file = "/content/drive/MyDrive/Neural_Signals_and_Computation_HW1/TEST_MOVIE_00001-small.tif"

# Load tif file into numpy array and image
data = tifffile.imread(file)

# Get mean, median, and variance
data_mean = np.mean(data, axis=0)
data_median = np.median(data, axis=0)
data_var = np.var(data, axis=0)

# Plot the summary images
# Create a square layout for the plot
layout = go.Layout(
    width=data.shape[1],  # Set the width of the plot
    height=data.shape[2],  # Set the height of the plot
    xaxis=dict(range=[0, data.shape[1]]),  # Set the x-axis range to match the width of the image
    yaxis=dict(range=[0, data.shape[2]]),  # Set the y-axis range to match the height of the image
    margin=dict(l=0, r=0, t=0, b=0),  # Set the margins to 0 to remove unnecessary spacing
)

# Plot the summary images
fig_mean = go.Figure(data=go.Heatmap(z=data_mean), layout=layout)
fig_median = go.Figure(data=go.Heatmap(z=data_median), layout=layout)
fig_var = go.Figure(data=go.Heatmap(z=data_var), layout=layout)

fig_mean.show()
fig_median.show()
fig_var.show()
```

The sumary plots for the mean and median of the data appear quite similar, showing a bright image with clear outlines for many of the cells visible in the image. Since both mean and median are statistic about centrality, they show similar results on the middle of the data flourescence information. On the other hand, the cells visible when we look at the variance of the data set, we can only observe roughly 5-7 cells bright enough to count. Thus, we can see a "distribution" of the variance within the data, indicating that only a couple of cells have enough variance in brightness (on and off?) to be seen in this summary statistic
